```python
slacAmps = {'C10':'AMP01','C11':'AMP02','C12':'AMP03','C13':'AMP04',\
           'C14':'AMP05','C15':'AMP06','C16':'AMP07','C17':'AMP08',\
           'C07':'AMP09','C06':'AMP10','C05':'AMP11','C04':'AMP12',\
           'C03':'AMP13','C02':'AMP14','C01':'AMP15','C00':'AMP16'}
slacAmps2 = {y: x for x, y in slacAmps.items()}
```

## R03-S12 (ITL)

```python
import sys
sys.path.append('/sdf/home/a/abrought/bin')
from SpotAnalysis import *
_, ptc_turnoffs_linearized, cti_turnoffs, moss, gains, gains_linearized= getSensorData(sensor="R03_S12", detector=23)
```

```python
# Get the gains on each amplifier in order to do background illumination subtraction
#Get the PTC collections
from lsst.daf.butler import Butler
from lsst.obs.lsst import LsstCam
sensor="R03_S12"
butler = Butler("/repo/ir2")
ptcs = butler.get('ptc', detector=23, instrument='LSSTCam', collections='u/abrought/BF/run_13144/ptcs_linearized')

camera = LsstCam().getCamera()
detector = camera.get("R03_S12")
    
# Get gains
gains = dict()
ptcturnoffs = dict()
for i, amp in enumerate(camera[0].getAmplifiers()):
    gains.update({amp.getName(): ptcs.gain[amp.getName()]})
    ptcturnoffs.update({amp.getName(): ptcs.ptcTurnoff[amp.getName()]})
```

```python
fig, axs = plt.subplots(2,8, sharex=True, sharey=True,figsize=(20,8))
for i in range(16):
    channel = slacAmps2[f"AMP{i+1:02d}"]
    ax = axs.flatten()[i]
    ax.axhline(0.0, linestyle="--", color="gray", linewidth=1)
    means = np.asarray(ptcs.rawMeans[channel]) * gains[channel]
    ax.plot(means,np.asarray(ptcs.covariances[channel])[:,0,1]/np.asarray(ptcs.covariances[channel])[:,0,0], "o", ms=2, label="Cov(0,1)")
    ax.plot(means,np.asarray(ptcs.covariances[channel])[:,1,0]/np.asarray(ptcs.covariances[channel])[:,0,0], "o", ms=2, label="Cov(1,0)")
    ax.set_title(f"AMP {i+1:02d}")
    ax.set_xlim(1000,200000)
    ax.set_xticks([10000,50000,100000,150000, 200000])
    ax.set_ylim(-0.02, 0.1)
    ax.axvline(ptc_turnoffs_linearized.__array__()[i], color="black", linestyle="--", alpha=0.5, label="PTC turnoff")
    ax.axvline(cti_turnoffs.__array__()[i], color="black", linestyle="-", alpha=0.5, label="sCTI turnoff")

# set labels
plt.setp(axs[-1, :], xlabel="$\mu$ [$e^{-}$]")
plt.setp(axs[:, 0], ylabel="$Cov(\;i,\:j\;)\;\;/\;\;\mu$")
#fmt = matplotlib.ticker.ScalarFormatter(useMathText=True)
#fmt.set_powerlimits((0, 0))
plt.ticklabel_format(axis="x", style="sci", scilimits=(0,0), useMathText=True)
axs.flatten()[0].legend(loc=2)
plt.savefig("../plots/" + sensor + "_covariances.png", bbox_inches='tight')
```

### Covariances and Model fit

```python
import matplotlib.pyplot as plt
import numpy as np
butler = dB.Butler("/repo/ir2", collections=["u/abrought/BF/run_13144/ptcs_linearized_fullcovariance_12_14_22"])
ptc = butler.get('ptc', instrument="LSSTCam", detector=112)

plt.figure(figsize=(10,8))
#fig, axs = plt.subplots(1,2, figsize=(20,8))
channel = 'C00'
means = np.asarray(ptc.finalMeans[channel]) * ptc.gain[channel]
plt.plot(means,np.asarray(ptc.covariances[channel])[:,0,1]/means, "o", c="blue", ms=2, label="Cov(0,1)")
plt.plot(means,np.asarray(ptc.covariancesModel[channel])[:,0,1]/means, "o-", c="blue", ms=2, label="Model")
plt.plot(means,np.asarray(ptc.covariances[channel])[:,1,0]/means, "o", c="orange", ms=2, label="Cov(1,0)")
plt.plot(means,np.asarray(ptc.covariancesModel[channel])[:,1,0]/means, "o-", c="orange", ms=2, label="Model")
#plt.plot(means,np.asarray(ptc.covariances[channel])[:,0,0]/means, "o", ms=2, label="Cov(0,0)")

#plt.xlim(1000,200000)
plt.ylabel("cov/mu")
#plt.plot(means, np.asarray(ptcs.rawVars[channel])/(1e6))
```

```python
import lsst.daf.butler as dB
butler = dB.Butler("/repo/ir2", collections=["u/abrought/BF/run_13144/ptcs_linearized_fullcovariance_12_14_22"])
ptc = butler.get('ptc', instrument="LSSTCam", detector=23)

fig, axs = plt.subplots(3,3, sharex=True, sharey=False,figsize=(15,15))
for x in range(3):
    for y in range(3):
        ax = axs[y][x]
        for i in range(16):
            channel = slacAmps2[f"AMP{i+1:02d}"]
            means = np.asarray(ptc.finalMeans[channel]) * ptc.gain[channel]
            ax.plot(means,(np.asarray(ptc.covariances[channel])[:,x,y] - np.asarray(ptc.covariancesModel[channel])[:,x,y]) / np.asarray(ptc.covariancesModel[channel])[:,x,y], "o", ms=2)
            ax.axhline(0.0, linestyle="--", color="gray", linewidth=1)
            ax.set_title("Cov(" + str(x) + "," + str(y) + ")")
            
plt.ticklabel_format(axis="x", style="sci", scilimits=(0,0), useMathText=True)
plt.suptitle("FIT to Astier+19 Eqn 20 (ITL)", fontsize=25)
plt.ylim(-1,1)
plt.ylabel("residual")
plt.xlabel("$\mu$ [e-]")
```
