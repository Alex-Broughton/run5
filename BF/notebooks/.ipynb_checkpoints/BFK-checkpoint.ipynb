```python
bfk1 = butler.get('bfk', detector=23, instrument='LSSTCam', collections='u/abrought/BF/2023.04.28/bfk.2023.04.28.R03-S12.trunc_to_pcti.amatrix.improved_corr_model')
#bfk2 = butler.get('bfk', detector=112, instrument='LSSTCam', collections='u/abrought/BF/2023.01.26/bfks.2023.01.26')
#bfk3 = butler.get('bfk', detector=det, instrument='LSSTCam', collections='u/abrought/BF/2023.01.26/bfks.2023.01.26.max100kE')
#bfk4 = butler.get('bfk', detector=det, instrument='LSSTCam', collections='u/abrought/BF/2023.01.26/bfks.2023.01.26.max100kE.noQuadFit')
#bfk5 = butler.get('bfk', detector=det, instrument='LSSTCam', collections='u/abrought/BF/2023.01.26/bfks.2023.01.26.noQuadFit.corrModelr_3')
#bfk6 = butler.get('bfk', detector=det, instrument='LSSTCam', collections='u/abrought/BF/2023.01.26/bfk.2023.01.26.max100kE.noAMP6')
#bfk7 = butler.get('bfk', detector=det, instrument='LSSTCam', collections='u/abrought/BF/2023.01.26/bfk.2023.01.26.max100kE.noAMP6.corrModel_r3')



def avg_kernels(d, exclude_channels=[]):
    t = np.zeros(d["C00"].shape)
    for k,channel in zip(d.values(), d.keys()):
        if channel in exclude_channels:
            continue
        t += k
    return t / len(d.values())

mu1 = avg_kernels(bfk1.ampKernels)
#mu2 = avg_kernels(bfk2.ampKernels, exclude_channels=['C01'])
#mu3 = avg_kernels(bfk3.ampKernels)
#mu4 = avg_kernels(bfk4.ampKernels)
#mu5 = avg_kernels(bfk5.ampKernels, exclude_channels=['C15'])
#mu6 = avg_kernels(bfk6.ampKernels, exclude_channels=['C15'])



fig, axs = plt.subplots(1,2, figsize=(16,8), facecolor='w')
plt.suptitle("Kernel profile for " + sensor, fontsize="x-large")   
axs[0].plot(mu1[:,8], drawstyle='steps-mid', label="itl", alpha=0.5)
#axs[0].plot(mu2[:,8], drawstyle='steps-mid', label="e2v", alpha=0.5)
#axs[0].plot(mu3[:,8], drawstyle='steps-mid', label="Max 100kE", alpha=0.5)
#axs[0].plot(mu4[:,8], drawstyle='steps-mid', label="Max 100kE + NoQuadFit", alpha=0.5)
#axs[0].plot(mu5[:,8], drawstyle='steps-mid', label="Max100kE \n+ No AMP6", alpha=0.5)
#axs[0].plot(mu6[:,8], drawstyle='steps-mid', label="Max100kE \n+ No AMP6 \n + CorrModelr3", alpha=0.5)

axs[1].plot(mu1[8,:], drawstyle='steps-mid', label="itl", alpha=0.5)
#axs[1].plot(mu2[8,:], drawstyle='steps-mid', label="e2v", alpha=0.5)
#axs[1].plot(mu3[8,:], drawstyle='steps-mid', label="Max 100kE", alpha=0.5)
#axs[1].plot(mu4[8,:], drawstyle='steps-mid', label="Max 100kE + NoQuadFit", alpha=0.5)
#axs[1].plot(mu5[8,:], drawstyle='steps-mid', label="Max100kE \n+ No AMP6", alpha=0.5)
#axs[1].plot(mu6[8,:], drawstyle='steps-mid', label="Max100kE \n+ No AMP6 \n + CorrModelr3", alpha=0.5)


axs[0].axhline(0.0,linestyle="--", color="black", alpha=0.5)
axs[1].axhline(0.0,linestyle="--", color="black", alpha=0.5)
axs[0].legend(loc=3)

axs[0].set_title("Parallel Slice")
axs[1].set_title("Serial Slice")
```
