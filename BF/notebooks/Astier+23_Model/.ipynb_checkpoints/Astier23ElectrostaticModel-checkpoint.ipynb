```python
def avg_kernels(d, exclude_channels=[]):
    t = np.zeros(np.asarray(d["C00"]).shape)
    for k,channel in zip(d.values(), d.keys()):
        if channel in exclude_channels:
            continue
        t += np.asarray(k)
    return t / len(d.values())

def get_sig_a(ptc):
    scaled = dict()
    for amp in ptc.covariances.keys():
        mask = np.asarray(ptc.expIdMask[amp])
        scale_factors = np.asarray(ptc.rawMeans[amp])**2
        arr = []
        for i, scale_factor in enumerate(scale_factors):
            arr.append(np.asarray(ptc.covariances[amp])[i] / scale_factor)
            
        scaled[amp] = np.std(np.asarray(arr)[mask], axis=0)

    covSqrtWeights = avg_kernels(scaled)

    return covSqrtWeights
```

```python
A = avg_kernels(ptc.aMatrix)
sig_a = get_sig_a(ptc)
```
