```python
import os
import numpy as np
import glob
from astropy.io import fits
import pickle as pkl
import matplotlib.pyplot as plt
from lsst.daf.butler import Butler
import lsst.afw.display as afwDisplay
import lsst.afw.image as afwImage
from astropy.table import Table, vstack, join
from scipy import stats
from scipy.optimize import curve_fit
from lsst.obs.lsst import LsstCam
camera = LsstCam().getCamera()



slacAmps = {'C10':'AMP01','C11':'AMP02','C12':'AMP03','C13':'AMP04',\
           'C14':'AMP05','C15':'AMP06','C16':'AMP07','C17':'AMP08',\
           'C07':'AMP09','C06':'AMP10','C05':'AMP11','C04':'AMP12',\
           'C03':'AMP13','C02':'AMP14','C01':'AMP15','C00':'AMP16'}
```

```python
repo_path = "/sdf/group/lsst/camera/IandT/repo_gen3/BOT_data/butler.yaml"
butler = Butler(repo_path)
registry = butler.registry

# Pick a sensor
# Interesting sensors: 9 (C15/C16), 31 (C10) 
sensor = "R24_S11"
# (9, 23, 31, 83, 112, 136)
det = 94
```

#### Read in BFKs

```python
bfk_r = butler.get('bfk', detector=det, instrument='LSSTCam', collections='u/abrought/BF/run_7068D/bfks_raw_rband')
bfk_r.ampKernels

fig, axs = plt.subplots(2,8, sharex=True, sharey=True, figsize=(20,8), facecolor='w')
plt.suptitle("Kernel profile for " + sensor, fontsize="x-large")   
for i, amp in enumerate(camera[0].getAmplifiers()):
    ax = axs[0 if i < 8 else 1][i if i < 8 else i-8]
    im = ax.plot(bfk_r.ampKernels[amp.getName()][:,8], drawstyle='steps-mid')
    im2 = ax.plot(bfk_r.ampKernels[amp.getName()][8,:], drawstyle='steps-mid')
    ax.axhline(0.0,linestyle="--", color="black", alpha=0.5)
    ax.set(title=amp.getName() + "/" + slacAmps[amp.getName()])
```

```python
bfk_i = butler.get('bfk', detector=det, instrument='LSSTCam', collections='u/abrought/BF/run_7069D/bfks_raw_iband')
bfk_i.ampKernels

fig, axs = plt.subplots(2,8, sharex=True, sharey=True, figsize=(20,8), facecolor='w')
plt.suptitle("Kernel profile for " + sensor, fontsize="x-large")   
for i, amp in enumerate(camera[0].getAmplifiers()):
    ax = axs[0 if i < 8 else 1][i if i < 8 else i-8]
    im = ax.plot(bfk_i.ampKernels[amp.getName()][:,8], drawstyle='steps-mid')
    im2 = ax.plot(bfk_i.ampKernels[amp.getName()][8,:], drawstyle='steps-mid')
    ax.axhline(0.0,linestyle="--", color="black", alpha=0.5)
    ax.set(title=amp.getName() + "/" + slacAmps[amp.getName()])
```

```python
bfk_g = butler.get('bfk', detector=det, instrument='LSSTCam', collections='u/abrought/BF/run_7067D/bfks_raw_gband')
bfk_g.ampKernels

fig, axs = plt.subplots(2,8, sharex=True, sharey=True, figsize=(20,8), facecolor='w')
plt.suptitle("Kernel profile for " + sensor + " TS8", fontsize="x-large")   
for i, amp in enumerate(camera[0].getAmplifiers()):
    ax = axs[0 if i < 8 else 1][i if i < 8 else i-8]
    im = ax.plot(bfk_g.ampKernels[amp.getName()][:,8], drawstyle='steps-mid', color="#6200FF", label="g-band")
    im2 = ax.plot(bfk_r.ampKernels[amp.getName()][:,8], drawstyle='steps-mid', color="#FF7200", label="r-band")
    im3 = ax.plot(bfk_i.ampKernels[amp.getName()][:,8], drawstyle='steps-mid', color="#870000", label="i-band")
    #im2 = ax.plot(bfk_g.ampKernels[amp.getName()][8,:], drawstyle='steps-mid')
    ax.axhline(0.0,linestyle="--", color="black", alpha=0.5)
    ax.set(title=amp.getName() + "/" + slacAmps[amp.getName()])
    if i==0:
        ax.legend(loc=3)
```
